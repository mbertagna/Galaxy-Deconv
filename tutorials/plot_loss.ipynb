```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../train_output_20250224_215239_loss.txt', header=None, names=['train_loss', 'val_loss'])
df.head()
```

```python
training_steps = list(range(df.shape[0]))


for col in list(df.columns):
    plt.plot(training_steps, df[col].values, label=col)

plt.title('')
plt.xlabel('training steps')
plt.ylabel('ellipse loss')
plt.legend()
plt.show()
```
